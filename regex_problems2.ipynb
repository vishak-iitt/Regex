Q.1. Group A: The Basics (Existence & Identification)
The Extension Filter: Match a filename. If it has a dot ., it must be followed by exactly 3 letters; otherwise, it must end in a digit.

```python
import re
pattern = r'^(.*?)(\.)?(?(2)[A-Za-z]{3}|\d)$'
text1="file.123"
text2="file"
print(re.findall(pattern,text1))
```

Q.3. The Bracket Balance: Match a string that starts with an optional (. If the ( was present, it must end with a ).

```python
import re
pattern = r"(?P<first>^\()(?(first).*\)$)"
text1="(abc)"
text2=""
print(re.findall(pattern,text1))
print(re.findall(pattern,text2))
```

```python
"""
Q.4 Group B: Pattern Logic (Conditional Branches)
The Area Code: Match a phone number. If it starts with a 
( (Group 1), it must match (\d{3})\d{3}-\d{4}. If no ( exists, 
it must match \d{10}.

"""

import re
pattern = r"(?P<first>^\()(?(first)(\d{3}\)\d{3}-\d{4})|(\d{10}))" 
#pattern = r"(?P<first>^\()(?(first)(\d{3}\)\d{3}-\d{4})|(\d{10}))"
text1 = "(123)456-7890" # With bracket
text2 = "1234567890"

print(re.findall(pattern,text1)) # [('(', '123)456-7890', '')]
print(re.findall(pattern,text2)) # []
```

```python
"""
Structural flaw

You intended:

“If the string starts with (, use one format, else use another”

But you wrote:

“Match ( at the start, and if that succeeded, use one format, 
else another”

Those are not equivalent.

The group is mandatory, so failure kills the entire match.
"""

"""
Q.4 Group B: Pattern Logic (Conditional Branches)
The Area Code: Match a phone number. If it starts with a 
( (Group 1), it must match (\d{3})\d{3}-\d{4}. If no ( exists, 
it must match \d{10}.

"""

import re
pattern = r"(?P<first>^\()?(?(first)(\d{3}\)\d{3}-\d{4})|(\d{10}))" 
text1 = "(123)456-7890" # With bracket
text2 = "1234567890"

print(re.findall(pattern,text1)) # [('(', '123)456-7890', '')]
print(re.findall(pattern,text2)) # []
```

```python
"""
Q.5. The Quote Closer: Match a string starting with either a 
single ' or double " quote. Use a conditional to ensure the 
closing quote matches the opening one.
"""

import re
pattern = r"^(['\"])(?(1).*(\1)$|)"
text = "'abc\""

print(re.findall(pattern,text))
```

```python
"""
Q.6. The Protocol Path: Match a URL. If it starts with https, it
must end with .gov. If it starts with http, it can end with
anything.

"""

import re
pattern = r"(^(https)?(?(1)(.*)\.gov$|))"
text = "https://www.higherpolynomial.gov"
#text = "www.higherpolynomial.math"

print(re.findall(pattern,text))
```

```python
"""
Q.7. The Protocol Path: Match a URL. If it starts with https, it
must end with .gov. If it starts with http, it can end with
.in. If it starts with www, it must end with .math

"""

import re
pattern = r"^(https)?(http)?(www)?(?(1)(.*)\.gov$|(?(2)(.*)\.in$|(?(3)(.*)\.math|)))"
text = "http://www.higherpolynomial.ini"
text = "www.higherpolynomial.math"

print(re.findall(pattern,text))
```

```python
"""
Q.8 The Currency Toggle: Match a number. If it starts with $, it 
cannot have a decimal. If it doesn't start with $, it must have 
exactly two decimal places."""

import re
pattern = r"^(\$)?(?(1)([1-9][0-9]*)|([0-9]+\.\d{2}))"
text = "$12"
text = "12.89"

print(re.findall(pattern,text))
```

```python
"""Q.9. Group D: "The Master Class" (Nested & Complex)
The Logic Gate: If Group 1 matches "Login", check for Group 2 
"Password". If Group 1 is "Guest", skip the password check.
"""

import re
pattern = r'^login:(?:(guest)(?!,password)|(\w+),password:(\w+))$'
pattern = r'^login:(?:(guest)|(\w+),password:(\w+))$'
text = "login:vishak,password:vishak"
text = "login:guest"

print(re.findall(pattern,text))

```

```python
"""
Q.10 — Conditional Branch (Prefix Control)
Match a string of the form type:value. If type is int, value must
be digits only. If type is str, value must be lowercase letters
only.

Examples:

int:12345 → match
str:hello → match
int:12a → reject
str:Hello → reject
"""

import re
pattern = r'^(?:(int:\d+)|(str:([a-z]+)))$'
pattern = r'^(?:(int:\d+)|(str:(?i:[a-z]+)))$'
text="str:Hello"
text="int:12345"
text="int:a12345"
print(re.findall(pattern,text))
```

```python
""" Q.11. — Optional Segment with Dependency
Match user:<name> optionally followed by ,id:<number>. If id 
appears, <name> must be at least 5 characters long.

Examples:

user:rohan → match
user:rohan,id:123 → match
user:raj,id:123 → reject
user:raj → match

"""

import re
pattern = r'^(user:\w{5,},id:\d+|user:\w+)$'
text="user:rohan,id:123"
print(re.findall(pattern,text))
```
