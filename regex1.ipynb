# Regular Expressions

There are 12 punctuation characters that are used in regex : 

$ ( ) * + . ? [ \ ^ { |

These 12 punctuation characters that make regular expressions work are called meta-characters. 

If you want your regex to match them literally, you need to escape them by using the backslash (\) in front of them. 


The punctuation characters, namely, ],-,} are not included in this set.

] and - become meta-characters only after an unescaped [

} becomes a metacharacter only after an unescaped {

There is no need to ever escape }



The symbols \  ^  - ] are the only ones that have special function inside a character class. Rest other need not be escaped when written inside the character class.

```python
# Match a correct date
# 23/11/1990
# 23-11-1990
# 23\11\1990

import re
text = "23-11-1989, 23\11\1989, 23/11/1989, 23.11.1989,23.11.1989a"

print(re.findall(r"\b(\d\d[\\\-/.]\d\d[\\\-/.]\d\d\d\d)\b", text))

# Remember that The symbols \  ^  - ] are the only ones that have special function inside a 
# character class. Rest other need not be escaped when written inside the character class.
```

**^**

Matches position at start of string.

With (?m), also matches immediately after each \n.


**$**

Matches position at end of string.

Also matches before a trailing newline.

With (?m), matches before every \n.


**\A**

Matches only at index 0.

Immune to all flags.


**\Z**

Matches only at the absolute end of the string

Does not match before a final trailing newline

Unaffected by (?m) and (?s)

# Word matching

In Python’s re module, \b is a zero-width assertion that matches a position where a word boundary occurs. It does not consume characters; it only asserts that the current position lies between:

a word character (\w: [A-Za-z0-9_])

and a non-word character (\W) or string boundary

# Formal Definition

Correct evaluation model

A position i satisfies \b iff:

is_word(text[i-1]) != is_word(text[i]) i.e. you are able to transition across a boundary

Apply to " cat "

[0] ' '   [1] 'c'   [2] 'a'   [3] 't'   [4] ' '

Position before c (between 0 and 1)

left = ' ' → \W → False

right = 'c' → \w → True

False != True → boundary exists

Position after t (between 3 and 4)

left = 't' → \w → True

right = ' ' → \W → False

True != False → boundary exists

Why your statement fails

“space is \W and hence not a \b”

This is a category error.

\W → character classification

\b → relationship between two adjacent characters

A space being \W is exactly what enables a boundary when next to a \w.

What actually happens

At both sides of "cat":

' ' | 'c'   → boundary

't' | ' '   → boundary

So \bcat\b matches.

Minimal model

Ignore characters. Evaluate positions only:

boundary exists ⇔ one side is \w and the other is \W

```python
import re

text = "cat scatter category cat"
print(re.findall(r"\bcat\b", text))# ['cat', 'cat']

text = "cat"
print(re.findall(r"\bcat\b", text)) # ['cat']
      
text = "scatter"
print(re.findall(r"\bcat\b", text)) # []      
```

```python
text = "category"
print(re.findall(r"\bcat\b", text)) # []

text = " cat "
print(re.findall(r"\bcat\b", text)) # ['cat']

print(re.findall(r"\bcat", "cat scatter category"))
# ['cat', 'cat']
```

```python
import re 

print(re.findall(r"\bcat", "cat")) # ['cat']

print(re.findall(r"\bcat", "scatter")) # []

print(re.findall(r"\bcat", "category")) # ['cat']
```

# Matching one of Several alternatives

There are no word boundaries while trying to match among several alternatives. The regex engine in this case (unless it has a word boundary \b) is eager i.e. it matches even if there is a partial match!

```python
import re 
print(re.findall(r"Mary|Jane|Suzy", "Mary, Jane, Sue")) # ['Mary', 'Jane']

print(re.findall(r"Mary|Jane|Suzy", "Mary, Janet, Sue")) # ['Mary', 'Jane']

print(re.findall(r"Mary|Jane|Suzy", "Mary, Janet, SuzyMary")) # ['Mary', 'Jane', 'Suzy', 'Mary']
```

```python
import re 
print(re.findall(r"Jane|Janet", "Her name is Janet")) # ['Jane']

print(re.findall(r"Janet|Jane", "Her name is Janet")) # ['Janet']

print(re.findall(r"\bJane|Janet", "Her name is Janet")) # ['Jane']

print(re.findall(r"\bJane\b|Janet", "Her name is Janet")) # ['Janet']
```

```python
import re

print(re.findall(r"\bMary|Jane|Sue\b", "Her name is Janet,MarySue")) # ['Jane', 'Mary', 'Sue']

print(re.findall(r"\bMary|Jane|Sue\b", "Her name is JanetMarySue")) # ['Jane', 'Sue']

print(re.findall(r"\b(Mary|Jane|Sue)\b", "Her name is Janet,MarySue")) # []
```

In the regular expression, " \b(\d\d)[\\\-/.](\d\d)[\\\-/.](\d\d\d\d)\b "

there are 3 capturing groups that are numbered from left to right. (\d\d) is group no 1 , (\d\d) is group no 2 and (\d\d\d\d) is group no 3. 

But \b(?:Mary|Jane|Sue)\b constitutes a non capturing group with the uwe of the flag ?:

You can also club it together with the case sensitive flag like this

```python
import re

text = "Her name is Janet,Mary, mary, Sue, sue"
print(re.findall(r"\b(?i:Mary|Jane|Sue)\b,", text)) # ['Mary,', 'mary,', 'Sue,']

text = "Her name is jane,Vishak"
print(re.findall(r"\b(?i:Mary|Jane|Sue)\b,\b(?i:vishak|vishal|lama)\b", text)) # ['jane,Vishak']
```

You used:

(?i)

in the middle of the pattern

Python’s re module does not allow global flags mid-pattern.

Formal rule (Python-specific)

(?i)

Allowed only at absolute start of the pattern

Otherwise → PatternError

Why your earlier reasoning failed

You assumed:

(?i) = toggle state anywhere

This is true in some regex engines (PCRE, etc.), but NOT in Python.

Python distinguishes:

Type	          Syntax     	Placement

Global flag	        (?i)	    ONLY at start

Scoped flag	      (?i:...)	    anywhere

Your failing pattern

\b(?:Mary|Jane|Sue)\b,\b(?i)(vishak|...)(?-i)\b,...

                        ↑ invalid position

Correct equivalent in Python

You must use scoped flags:

r"\b(?:Mary|Jane|Sue)\b,\b(?i:(vishak|vishal|lama))\b,\b(ramya|dhanya)\b"

Important consequence

(?i) → global, restricted placement

(?i:...) → local, unrestricted

About (?-i)

Same restriction applies:

Cannot be used mid-pattern as a global toggle

Only valid inside scoped form like:

(?i:... (?-i:...) ...)

?i: will just cuse the the second group to be case insensitive. 
what if i also want it not to be captured?

```python
import re

text = "Her name is Jane,Vishak,ramya"

print(re.findall(r"\b(?:Mary|Jane|Sue)\b,\b(?i:(?:vishak|vishal|lama))\b,\b(ramya|dhanya)\b",text))
# Match found : Jane,Vishak,ramya but Captured portion: ['ramya']

#alternately
print(re.findall(r"\b(?:Mary|Jane|Sue)\b,\b(?i:vishak|vishal|lama)\b,\b(ramya|dhanya)\b",text))
# Match found : Jane,Vishak,ramya but Captured portion: ['ramya']
```

# Backreferencing

Q : Assume that you don't have to exclude dates like : 9999-99-99 , create a regular expression that matches "magical" dates in yyyy-mm-dd format. A date is magical if the year minus the century, the month and the day of the month are all the same numbers.

```python
text = "2012-12-12"
print(re.findall(r"\b\d\d(\d\d)[-\\./]\1[-\\./]\1\b",text)) # ['12']

print(re.findall(r"\b(?P<year>\d\d\d\d)([-\\.\-/])(?P<month>\d\d)\2(?P<day>\d\d)\b",text))
```

```python
import re 
text = "2012-12-12"
m = re.search(
    r"\b(?P<year>\d{4})(?P<sep>[-\\./])(?P<month>\d{2})\2(?P<day>\d{2})\b",
    text
)

print(m.group('year'))
print(m.group('month'))
print(m.group('day'))
print(m.group('sep'))
```

# Quantifiers

A floating point number with an optional integer part, a mandatory fractional part, and an optional exponent. Each part allows any number of digits

```python
import re 
text= ".14"
print(re.findall(r"\b(\d)?\.(\d)+\b",text))
```

\d{5} --- match a digit exactly 5 number of times. 

\d{1,3} --- match a digit either 1 or 2 or 3 number of times.

ab{0}c --- same as ac

```python
# Match a 8 bit hexadecimal number

import re 
text= ".14"
print(re.findall(r"(\d)?\.(\d)+",text)) # [('', '4')]
#Python captures only the last iteration in the single captured group

print(re.findall(r"(\d)?\.(\d+)", text)) # [('', '14')]
```

Pattern: (\d)?\.(\d)+

Text: .14

Execution trace under Python’s re:

Start at index 0 (.)

(\d)? → optional → does not match (since . is not a digit) → group 1 = ''

\. → matches .

(\d)+ → greedy repetition:

matches '1'

matches '4'

stops (end of string)

Captured groups behavior

Group 1: optional, unmatched → ''

Group 2: (\d)+ → only last iteration is retained → '4'

Result

[('', '4')]

Key mechanism:

In Python regex, a capturing group inside a quantified pattern ((...) +) stores only the 

final iteration, not the full matched span.
