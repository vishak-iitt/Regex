# Regular Expressions

There are 12 punctuation characters that are used in regex : 

$ ( ) * + . ? [ \ ^ { |

These 12 punctuation characters that make regular expressions work are called meta-characters. 

If you want your regex to match them literally, you need to escape them by using the backslash (\) in front of them. 


The punctuation characters, namely, ],-,} are not included in this set.

] and - become meta-characters only after an unescaped [

} becomes a metacharacter only after an unescaped {

There is no need to ever escape }



The symbols    \  ^  - ]    are the only ones that have special function inside a character class. Rest other need not be escaped when written inside the character class.

```python
# Match a correct date
# 23/11/1990
# 23-11-1990
# 23\11\1990

import re
text = "23-11-1989, 23\11\1989, 23/11/1989, 23.11.1989,23.11.1989a"

print(re.findall(r"\b(\d\d[\\\-/.]\d\d[\\\-/.]\d\d\d\d)\b", text))

# Remember that The symbols \  ^  - ] are the only ones that have special function inside a 
# character class. Rest other need not be escaped when written inside the character class.
```

# Word matching

In Python’s re module, \b is a zero-width assertion that matches a position where a word boundary occurs. It does not consume characters; it only asserts that the current position lies between:

a word character (\w: [A-Za-z0-9_])

and a non-word character (\W) or string boundary

# Formal Definition and Correct evaluation model

A position i satisfies \b iff:

is_word(text[i-1]) != is_word(text[i]) i.e. you are able to transition across a boundary

Apply to " cat "

[0] ' '   [1] 'c'   [2] 'a'   [3] 't'   [4] ' '

Position before c (between 0 and 1)

left = ' ' → \W → False

right = 'c' → \w → True

False != True → boundary exists

Position after t (between 3 and 4)

left = 't' → \w → True

right = ' ' → \W → False

True != False → boundary exists

**Why your statement fails**

“space is \W and hence not a \b”

This is a category error.

\W → character classification

\b → relationship between two adjacent characters

A space being \W is exactly what enables a boundary when next to a \w.

**What actually happens**

At both sides of "cat":

' ' | 'c'   → boundary

't' | ' '   → boundary

So \bcat\b matches.

**Minimal model**

Ignore characters. Evaluate positions only:

boundary exists ⇔ one side is \w and the other is \W

```python
import re

text = "cat scatter category cat"
print(re.findall(r"\bcat\b", text))# ['cat', 'cat']

text = "cat"
print(re.findall(r"\bcat\b", text)) # ['cat']
      
text = "scatter"
print(re.findall(r"\bcat\b", text)) # []      
```

# Matching one of Several alternatives

There are no word boundaries while trying to match among several alternatives. The regex engine in this case (unless it has a word boundary \b) is eager i.e. it matches even if there is a partial match!

```python
import re 
print(re.findall(r"Mary|Jane|Suzy", "Mary, Jane, Sue")) # ['Mary', 'Jane']

print(re.findall(r"Mary|Jane|Suzy", "Mary, Mary, Mary")) # ['Mary', 'Mary', 'Mary']

print(re.findall(r"Mary|Jane|Suzy", "Mary, Janet, Sue")) # ['Mary', 'Jane']

print(re.findall(r"Mary|Jane|Suzy", "Mary, Janet, SuzyMary")) # ['Mary', 'Jane', 'Suzy', 'Mary']
```

```python
"""
Engine behavior (leftmost-first)

At each position:

1. Try alternatives in order (Mary → Jane → Suzy)

2. First successful one is chosen

3. That match is returned

4. Pointer moves to end of that match

Step-by-step

Input
Mary, Janet, SuzyMary
^
1) At index 0
Mary...
Try "Mary" → matches
Accept immediately (no need to try others)

Match:

Mary

Pointer moves to index 4.

2) Continue scanning
, Janet, SuzyMary
  ^

At index 6:

Janet
^
Try "Mary" → fail
Try "Jane" → matches first 4 letters of "Janet"
Accept

Match:

Jane

Pointer moves to after "Jane".

3) Continue
t, SuzyMary
   ^

At index 12:

SuzyMary
^
Try "Mary" → fail
Try "Jane" → fail
Try "Suzy" → match

Match:

Suzy

Pointer moves forward.

4) Continue
Mary
^
Try "Mary" → match

Match:

Mary
Final output
['Mary', 'Jane', 'Suzy', 'Mary']
Important observations
1) Partial matches are allowed
"Jane" matches inside "Janet"

No word boundary constraint.

2) Order matters
Mary|Jane|Suzy

If two alternatives could match at same position:

FIRST one wins
3) Non-overlapping

After matching:

pointer jumps to end of match

No overlapping results.

Core invariant
Alternation:
    tries patterns left-to-right
    returns first successful match at each position
    """
```

```python
import re 
print(re.findall(r"Jane|Janet", "Her name is Janet")) # ['Jane']

print(re.findall(r"Janet|Jane", "Her name is Janet")) # ['Janet']

print(re.findall(r"\bJane|Janet", "Her name is Janet")) # ['Jane']

print(re.findall(r"\bJane\b|Janet", "Her name is Janet")) # ['Janet']
```

# Greedy quantifiers

As the name suggests , greedy quantifiers begin my matching whole of the available text that IS PERMITTED. They then backtrack if the next pattern is not matched and see if a match is now made? if not, it then backtracks again until a match is found. 

```python
import re

text = "a\nb\nc"

print(re.findall(r".*", text)) # ['a', '', 'b', '', 'c', '']

print(re.findall(r"(?s).*", text)) # ['a\nb\nc', '']

print(re.findall(r"(?m).*", text)) # ['a', '', 'b', '', 'c', '']
# (?m) does not affect .
# It only changes behavior of ^ and $
#(?m) does not affect .
#It only changes behavior of ^ and $

print(re.findall(r"(?sm).*", text)) # ['a\nb\nc', '']
```

# Lazy quantifier

You can make any quantifier lazy by placing a question mark after it. Unlike greedy which consumes the whole of the available text that is permitted under the given operator, the lazy quantifier just moves one step at a time.

If we use "\d+?\b" on "1234", then "\d+?" first matches only 1. \b fails between 1 and 2; then \d+? expands to 12 and \b still fails. It continues till \d+? matches 1234 and then it encounters a word boundary and thus \b succeeds

Note that the + operator is greedy by default. If we use "\d+\b" on "1234", \d+ will match all the digits. \b then matches and then an overall match is found. 

If we use "\d+\b" on "1234X", then \d+ greedy matches whole of 1234 but then \b fails. Then \d+ backtracks to 123 and \b still fails. This continues until \d+ has backtracked to it's minimum 1 and \b still fails. Then the whole match attempt fails. 

If we use "\d+?\b" on "1234x", then first it matches only 1 but \b fails between 1 and 2. \d+? expands to 12 and \b still fails. This continues until \d+? matches 1234 and \b still fails. The regex engine attempts to expand \d+? once more but \d does not match x. The overall match attempt fails.

```python
import re

text = "a\nb\nc"

print(re.findall(r".", text)) # ['a', 'b',  'c']

print(re.findall(r"(?s).", text) ) # ['a', '\n', 'b', '\n', 'c']
```

# Possessive Quantifier

Possessive quantifiers eats the whole of the text initially, just like greedy quantifier. The difference is that possessive quantifier will never give back (i.e. it never backtracks), not even when giving back is the only way that the remainder of the regular expression could match. possessive quantifiers do not keep backtracking positions. 

You make any quantifier possessive by placing a ++ sign after it. . Python explicity does not have possessive quantifiers. However, it does support wrapping a greedy quantifier inside an atomic group. and that has the same effect as using a possessive quantifier

**Atomic group:** Syntax is (?> .....) where the ..... signify other regular expressions. The Atomic group is a non-capturing group that refuses to backtrack.  For example: 


```python
"""
(?>.*?</body>) throws away all the matching positions for 
.*?</body> when /body is found.



\d++ is the same as (?>\d+)

therefore, \bd++\b is the same as \b(?>\d+)\b


Suppose you apply \bd++\b to 123abc 456. \b matches at the word 
boundary at the start of the subject and \d++ matches 123; this is
exactly what a greedy quantifier would have also done. But, the 
second \b fails to match between 3 and a. But the possessive 
quantifer did not store any backtracking positiions. and thus, 
there are no further options to try. The regex engine immediately
declares failure. 
"""
```

```python
import re

text = "abc."

print(re.findall(r"\b(?>\w+)\b", text)) # ['abc']

text = "abc123."

print(re.findall(r"\b(?>\w+\d+)\b", text)) # ['abc123']

print(re.findall(r"\b(?>\w+)(?>\d+)\b", text)) # []

print(re.findall(r"\b(\w+)(?>\d+)\b", text)) # ['abc12']
```

```python
"""
Query : r"\b(?>\w+\d+)\b" shoul;d have consumed the whole of abc123
 as w+ is greedy and then because of atomic group it does not
   backtrack and then it cannot find a \d+ and it should have 
   resulted in a failure match
"""
```

```python
"""
Explanation: 

Actual engine behavior

Regex engines are leftmost-first, greedy, with backtracking (unless blocked).

Inside:

(?>\w+\d+)

execution still proceeds sequentially:

Step-by-step (real behavior)

Input:

abc123.

Step 1 — \w+ (greedy)

\w+ → consumes "abc123"

Step 2 — \d+

\d+ → fails (nothing left)

Step 3 — Backtracking inside atomic group

This is the crucial point:

Atomic group does NOT forbid internal backtracking while the group 
is still being evaluated

So engine tries:

\w+ → "abc12"
\d+ → "3"

Now:

success
Step 4 — Atomic lock

After success:

(?>...) locks the result
→ no further backtracking allowed OUTSIDE the group
Why Case 3 fails
(?>\w+)(?>\d+)

Now separation matters:

Step 1
(?>\w+) → consumes "abc123"
→ LOCKED
Step 2
(?>\d+) → fails

Now:

cannot backtrack into previous atomic group
→ total failure
Correct conceptual model
Atomic group:
    allows backtracking DURING its own evaluation
    forbids backtracking AFTER it has matched
What the explanation got wrong

It claimed:

engine “assigns” \w+ → abc directly

Reality:

engine first overshoots (abc123),
then backtracks internally,
then settles on abc12 + 3
Why your intuition seemed correct earlier

Your earlier reasoning:

\w+ will consume everything → \d+ fails → atomic prevents backtracking

was missing this nuance:

Atomic does NOT freeze intermediate greedy steps,
only the FINAL accepted match
Final invariant
Inside (?>X):
    backtracking is allowed while matching X

After (?>X) succeeds:
    X becomes immutable (no further backtracking into it)

"""
```

```python
import re

text = "abc123"

print(re.findall(r"(?>\w+\d+)\w", text))   # []
print(re.findall(r"(\w+\d+)\w", text))     # ['abc12']
```

Q. USe a single regular expression to match a complete HTML file, checking for properly nested html, head, title and body tags. 

```python
import re
html_file = "";

# NAIVE Solution
print(re.findall(r"(?s)<html>.*?<head>.*?<itle>.*?</title>" \
"                 .*?</head>.*?<body[^>]*>.*?</body>.*?</html>"
                  ,html_file))
```

The above code works perfectly fine on a proper HTML file. But the regex gets you into trouble when it deals with a subject text that does not have all the HTML tags. The worst case occurs when </html> is missing.

Imagine the regex engine has matched all preceeding tags and is now busy expanding the last 
.\*?. Since /html is missing and can never match, the just preceeding .\*? expands all the way to the end of the file. When it can no longer expand, it fails because /html is never found. 

But the other six ".\*?" have all remembered a backtracking position that allows them to expand further. When the last ".\*?" fails, the one before that also expands gradually matching 
/body . This .\*? will also expand all the way to the end of the file, as well all preceeding lazy dots. Only when the first one reaches the end of the file will the regex engine declare failure. 

This has a worst case scenario of n^7. WE call this  catastrophic backtracking. 

**SOLUTION** : USE Atomic grouping. 

```python
# EFFECIENT SOLUTION

import re
html_file = "";

# NAIVE Solution
print(re.findall(r"(?s)<html>(?>.*?<head>)(?>.*?<title>)\
                 (?>.*?</title>)(?>.*?</head>)(?>.*?<body[^>]*>)\
                 (?>.*?</body>).*?</html>", 
                 html_file))

# NOTE : [^>]* → any chars except > repeated 0 or more number of 
# times
```

They affect matching, not what is included in the match.

Types

Lookarounds do not advance or reposition the engine; they only assert a condition at the current index and leave the position unchanged.

Think of the regex "cursor" like the blinking cursor in a text editor. Normally, when a regex matches a character (like [A-Z]), the cursor moves past it to the next character. This is called consuming the text.

Lookarounds are different: they are "Zero-Width Assertions."

The "Ghost" Cursor
When the regex engine hits a lookaround, it pauses at its current position, "teleports" ahead to see if the condition is true, and then teleports back to the exact same spot before the rest of the pattern continues.

A Step-by-Step Breakdown
Imagine we are testing the string A1 against the pattern (?=[A-Z])[0-9]:

Start: The cursor is at the very beginning of the string (index 0), before the letter A.

The Lookahead (?=[A-Z]): The engine peeks ahead. Is there an uppercase letter at the current position? Yes (A).

The Reset: Because it's a lookaround, the cursor does not move past the A. It stays at index 0.

The Next Step [0-9]: Now the engine tries to match a digit starting from the current cursor position (index 0).

The Failure: It sees A, which is not a digit. The match fails.

The behavior of the cursor in a Negative Lookahead is exactly the same as in a Positive Lookahead: it does not move



```python
"""
Execution model

At any index i, the engine does:

Check (?=\d) → “is the character at position i a digit?”
If yes → run (\w+) starting at the same position i
Apply to "abc1 def ghi24 xyz"

At index of '1':

Step 1: (?=\d) → current char = '1' → condition satisfied

Step 2: (\w+) starts at '1'

'1' is a word character → consumed
next is space → stop

Result: "1"

At index of '2':

Step 1: (?=\d) → current char = '2' → satisfied

Step 2: (\w+):

'2' → consumed
'4' → consumed (still \w)
next is space → stop

Result: "24"

Critical point

The lookahead is evaluated before any characters are consumed, and it is anchored at the current position, not after \w+.

So the pattern enforces:

“start matching at a digit, then consume a word”

That is why "1" is captured.
"""
```

does the cursor move within a lookahead and then after it finishes it's job, it comes back to its starting position

The short answer is no, the cursor (or engine position) never actually "moves" in the traditional sense.

In a lookahead, the engine checks the text ahead without consuming any characters. It is often called a zero-width assertion.

How it Works Internally
Think of a lookahead like a scout standing at the current position. The scout runs forward to see if the pattern matches, but the "army" (the actual engine cursor) stays exactly where it is.

The Assertion: When the engine encounters a lookahead like (?=...), it pauses at the current index.

The "Virtual" Scan: It temporarily looks at the characters following that index to see if they satisfy the lookahead condition.

The Result: * If it matches, the engine stays at the same index it was at before the lookahead started and moves to the next part of your regex.

If it fails, the entire match attempt at that position fails (or triggers backtracking).

A Practical Example
Consider the regex: q(?=u)i
And the string: "quit"

Step 1: The engine matches q. The cursor is now between q and u.

Step 2: It sees the lookahead (?=u). It "peeks" at the next character. Is it a u? Yes.

Step 3: Because it was a lookahead, the cursor remains between q and u. It does not "land" on the u.

Step 4: The engine now tries to match the next part of the regex, which is i, against the next character in the string, which is u.

Result: This regex will fail to match "quit" because after the lookahead, the engine is looking for an i where the u actually is.

just like lookahead where the cursor scans and returns back, in lookbehind as well, the cursor should scan behind and come back to the orginal position right?



1. Lookahead (?=...)

The Engine stands at a position.

The Scout starts at that same position and runs forward (right).

The Return: The scout returns to the engine's position.

2. Lookbehind (?<=...)

The Engine stands at a position.

The Scout "jumps" a specific number of characters backward (left) and then scans forward toward the engine.

The Return: The scout returns to the engine's position.

Why the "Fixed-Width" rule exists

This is the "Aha!" moment for why lookbehinds are more restricted than lookaheads in many languages (like Python):


Lookahead: The engine says, "Scout, go right. I don't care how far you go, just tell me if you find a digit." The scout can run forever until it hits the end of the string.

Lookbehind: The engine says, "Scout, go left." But the scout needs to know exactly how many steps to jump back before it starts scanning forward to see if the match hits the engine's current spot.

If you say (?<=.*)abc, the engine doesn't know if it should jump back 1 character, 10 characters, or 1,000 characters to start checking. This is why Python's re module requires lookbehinds to have a fixed length (e.g., (?<=123)abc).





# Conditionals

# 1) Definition

A conditional in regex allows branching based on whether a group has matched or whether a lookaround condition is true.

**Conditionals are non capturing groups.**

General form:

(?(condition)yes-pattern|no-pattern)

**2) Condition types**

(A) Group existence condition

(?(1)yes|no)

```python
"""
While the "if-test" itself (the part that checks if a group exists) is zero-width and takes no time or space, the patterns you put inside the "Yes" and "No" branches behave exactly like normal regex patterns.

If a branch matches \d{3}, the regex cursor will move forward by three digits.

The Anatomy of Consumption
In the structure (?(1)ABC|XYZ):

The Test (?(1): Zero-width. The cursor stays still while the engine checks if Group 1 was captured.

The "Yes" Branch ABC: If Group 1 exists, the engine tries to match ABC. If it succeeds, the cursor moves past ABC.

The "No" Branch XYZ: If Group 1 doesn't exist, the engine tries to match XYZ. If it succeeds, the cursor moves past XYZ.
"""
```

```python
"""
When the regex engine evaluates (?(1)yes|), it sets up a fork in
the road:Condition Met (Group 1 exists): * The engine takes the 
yes branch. It works just like a normal part of your regex.

Condition Not Met (Group 1 is empty): * The engine takes the no 
branch.Because the branch is empty, it consumes $0$ characters. 
It doesn't throw an error and doesn't fail the overall match—it 
simply steps over the conditional block and continues processing 
whatever comes next in the pattern.
"""
```

```python
"""
Positive lookahead / lookbehind

Capturing groups do capture and are usable outside the lookaround.

Mechanism: a positive lookaround asserts that a subpattern does 
match at a position. Since the engine actually matches that 
subpattern (without consuming input), any capturing groups inside
it are populated.

Example:

import re

text = "abc123"
m = re.search(r'(?=(\w+))', text)

print(m.group(1))  # abc123

The lookahead (?=(\w+)) succeeds at position 0, and group 1 
contains the full match of \w+.

Another example with constrained scope:

m = re.search(r'\d(?=(\d{2}))', "1234")
print(m.group(1))  # 23

The engine matches 1, then asserts that the next two digits exist
and captures them.

Conclusion:
Positive lookarounds populate capture groups normally.

Negative lookahead / lookbehind

Capturing groups inside do not produce usable captures.

Mechanism: a negative lookaround asserts that a subpattern does 
not match. For the assertion to succeed, the inner pattern must 
fail. Since no successful match occurs inside, capturing groups 
have no defined content.

Example:

m = re.search(r'(?!a(b))', "ac")
print(m.group(1))  # None

Even though (b) exists syntactically, it never captures because 
the engine only succeeds when a(b) fails.

More subtle case:

m = re.search(r'(?!.*(a))', "bbb")
print(m.group(1))  # None

The lookahead succeeds because .*(a) fails, so (a) never captures.

Conclusion:
Negative lookarounds cannot yield meaningful captured data.

Summary
Positive lookaround: captures are valid and accessible.
Negative lookaround: captures are undefined / None because the 
inner match must fail.
"""
```
